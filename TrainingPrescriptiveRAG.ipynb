```python
df = pd.read_csv("running_data.csv")
```

```python
columns_to_keep = [
    'age', 'gender', 'vo2_max', 'hrv', 'lactate_threshold',
    'resting_heart_rate', 'max_heart_rate',
    'endurance_score', 'efficiency_score', 'adaptability_score'
]
df = df[columns_to_keep]
```

```python
df = df.dropna()
```

```python
def generate_prompt(row):
    return f"""
User Profile:
- Age: {row['age']}
- Gender: {row['gender']}
- VO2 Max: {row['vo2_max']}
- Lactate Threshold: {row['lactate_threshold']}
- Heart Rate Variability (HRV): {row['hrv']}
- Resting Heart Rate: {row['resting_heart_rate']}
- Max Heart Rate: {row['max_heart_rate']}
- Endurance Score: {row['endurance_score']}
- Efficiency Score: {row['efficiency_score']}
- Adaptability Score: {row['adaptability_score']}

Based on the above physiological profile, what type of training would you recommend?
(Options: Light Cardio, Endurance Cardio, Interval Training, HIIT, Strength Training)
"""

df['prompt'] = df.apply(generate_prompt, axis=1)
```

```python
def recommend_training(row):
    if row['vo2_max'] > 50 and row['endurance_score'] > 7:
        return "Endurance Cardio: Great aerobic fitness, focus on long sessions."
    elif row['hrv'] < 30:
        return "Light Cardio: Allow recovery due to low heart rate variability."
    elif row['adaptability_score'] > 5:
        return "Interval Training: Suitable for improving adaptability."
    else:
        return "Strength Training: Build base strength for performance."
```

```python
df['training_recommendation'] = df.apply(recommend_training, axis=1)
print(df[['age', 'gender', 'training_recommendation']])
```
